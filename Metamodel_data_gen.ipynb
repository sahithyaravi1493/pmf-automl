```python
# Imports
import pandas as pd
import json
import zipfile
import ast
import numpy as np
from pandas import HDFStore
```

```python
zf = zipfile.ZipFile('all_normalized_accuracy_with_pipelineID.zip') 
df = pd.read_csv(zf.open('all_normalized_accuracy_with_pipelineID.csv'))
df.head()
```

```python
features = pd.read_csv('data_feats_featurized.csv')
features.head()
```

```python
with open('pipelines.json', 'r') as f:
    pipelines = json.load(f)
```

```python
pipeline_dict = dict()
for pipeline in pipelines:
    pipe = dict(pipeline)
    del pipe['id']
    pipeline_dict[pipeline['id']] = pipe
```

```python
pipelines =[]
df2 = df
for index,row in df.iterrows():    
    pipelines.append((pipeline_dict[row['Unnamed: 0']]))
df2['pipeline'] = pipelines
```

```python
df1 = df.copy()
df1.set_index('Unnamed: 0',inplace=True)
df_changed = df1.stack().reset_index()
df_changed.head()
```

```python
df_changed.columns =['pipeline', 'Unnamed: 0', 'accuracy']
df_changed.head()
```

```python
df_changed['Unnamed: 0'] = df_changed['Unnamed: 0'].astype(int)
```

```python
res = pd.merge(df_changed, features, on='Unnamed: 0', how='left')
```

```python
res.shape
```

```python
res.to_pickle('df_id.pkl')
```

```python
res = pd.read_pickle('df_id.pkl')
pipeline_res=[]
pipeline_ids = np.array(res['pipeline'].values)
pipeline_ids.shape
```

```python
names = np.vectorize(pipeline_dict.get)(pipeline_ids)
```

```python
res['pipeline_names'] = names
```

```python
res.head()
```

```python
import pandas as pd
from pandas.io.json import json_normalize

df_2 = json_normalize(res['pipeline_names'])
```

```python
result = pd.concat([res, df_2], axis=1).drop('pipeline_names', axis=1)
```
